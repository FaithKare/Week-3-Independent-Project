```
#loading the telecom 1 dataset
dt1 = pd.read_csv('Telcom_dataset.csv')
dt1
dt1.columns =['PRODUCT', 'VALUE', 'DATETIME', 'CELLONSITE', 'DWANUMBER', 'DWBNUMBER', 'COUNTRYA', 'COUNTTRYB' , 'CELLID', 'SITEID']
dt1.head()
```

```
#loding telecom 2 dataset
dt2 = pd.read_csv('telcom_dataset2.csv')
dt2
dt2.columns =['PRODUCT', 'VALUE', 'DATETIME', 'CELLONSITE', 'DWANUMBER', 'DWBNUMBER', 'COUNTRYA', 'COUNTTRYB' , 'CELLID', 'SITEID']
dt2.head()
```

```
#loading telcom 3 dataset
dt3 = pd.read_csv('Telcom_dataset3.csv')
dt3
dt3.columns =['PRODUCT', 'VALUE', 'DATETIME', 'CELLONSITE', 'DWANUMBER', 'DWBNUMBER', 'COUNTRYA', 'COUNTTRYB' , 'CELLID', 'SITEID']
dt3.head()
```

```
#loading cells geographric dataset
dt4 = pd.read_csv("cells_geo.csv")
dt4.columns = ["mycolumns"]

#splitting the columns in the data 
dt4 = dt4.mycolumns.str.split(";",expand=True)
dt4
#renaming columns of the dataset
dt4.columns = ['INDEX','VILLES', 'STATUS', 'LOCALISATION', 'DECOUPZONE' ,'ZONENAME', 'LONGITUDE', 'LATITUDE', 'REGION', 'AREA', 'CELLID' ,'SITEID']
dt4.head()
```

```
#cleaning the first dataset
#checking for missing values
dt1.isnull()
 
#dropping duplicated values
dt1.duplicated()
dt1.drop_duplicates()

#droping rows if all cells are missing 
dt1_cleaned = dt1.dropna(how='all')
dt1_cleaned

#droping columns if all cells are missing 
dt1.dropna(axis=1, how='all')

dt1.dropna(thresh=10)
```

```
#cleaning the second dataset
#checking for missing values
dt2.isnull()

#checking and dropping duplicated values
dt2.duplicated() 
#dropping duplicated values
dt2.drop_duplicates()

#droping rows if all cells are missing 
#dt2_cleaned = dt1.dropna(how='all')
#dt2_cleaned

#droping columns if all cells are missing 
#dt2.dropna(axis=1, how='all')

dt2.dropna(thresh=10)
```

```
#cleaning the  telecom3 dataset
#checking for missing values
dt3.isnull()

#dropping duplicated values
dt3.duplicated()
dt3.drop_duplicates()

#droping rows if all cells are missing 
dt3_cleaned = dt1.dropna(how='all')
dt3_cleaned

#droping columns if all cells are missing 
dt3.dropna(axis=1, how='all')

dt3.dropna(thresh=10)
```

```
#claening the cells geogrphical data
#checking for missing values
dt4.isnull()

#dropping duplicated values
dt4.drop_duplicates()
dt4_no_missing = dt4.dropna()
dt4_no_missing

#droping rows if all cells are missing 
dt4_cleaned = dt4.dropna(how='all')
dt4_cleaned

#droping columns if all cells are missing 
dt4.dropna(axis=1, how='all')

dt4.dropna(thresh=12)
```

```
#merging the three telecom datasets
dt5 = pd.concat((dt1,dt2,dt3),axis =0,ignore_index=True)
dt5.head()
```

```
#merging the mergerd telecom dataset with the cells geo
dt6 = dt5.merge(dt4, on='SITEID', how='inner')

#dropping columns COUNTRYA, COUNTRYB and CELL ON SITEas theywould not be usesful in analysis
dt6.drop(columns=['COUNTRYA','COUNTTRYB', 'CELLONSITE'],axis=1, inplace=True)
dt6.head()
```

```
#converting the date time column from string to datetime objects

dt6['DATETIME'] = pd.to_datetime(dt6['DATETIME'])
dt6['DATETIME']

#to get the cities most used during business and home , we calculate the sum of datetime hours for every city
dt6.groupby([dt6['DATETIME'].dt.hour, 'VILLES'])['VALUE'].sum()

```
